```python
class MatchRow:
    def __init__(self, home, away, goal_1, goal_2, week):
        self.home = home
        self.away = away
        self.goal_1 = goal_1
        self.goal_2 = goal_2
        self.week = week
        self.result = 0
        if (self.goal_1 > self.goal_2):
            self.result = 1
        elif (self.goal_2 > self.goal_1):
            self.result = 2
```

```python
ev = []
depl = []
goals_ev = []
goals_depl = []
session = requests.Session()

for w in range(1,35):
    url = "https://www.tff.org/Default.aspx?pageID=1413&hafta=" + str(w) + "#grp"
    response = session.get(url)
    print("W: ", w, "\n")
    içerik = response.content
    soup = BeautifulSoup(içerik, "html.parser")
    str1 = "ctl00_MPane_m_1413_9320_ctnr_m_1413_9320_dtlHaftaninMaclari_ctl"
    str2 = "_Label"
 
    for i in range(1,10,1):
        if len(str(i)) == 1:
            match_id = str1 + "0" + str(i) + str2
            goal_id = str1 + "0" + str(i) + str2
        else:
            match_id = str1 + str(i) + str2
            goal_id = str1 + str(i) + str2

        match_id1 = match_id + "4"
        match_id2 = match_id + "1"
        goal_id1 = goal_id + "5"
        goal_id2 = goal_id + "6"
        result_ev = soup.find_all("span",{"id": match_id1})
        result_depl = soup.find_all("span",{"id": match_id2})
        goal_1 = soup.find_all("span", {"id": goal_id1})
        goal_2 = soup.find_all("span", {"id": goal_id2})

        for x in result_ev:
            ev.append(x.getText())
        for x in result_depl:
            depl.append(x.getText())
        for x in goal_1:
            goals_ev.append(x.getText())
        for x in goal_2:
            goals_depl.append(x.getText())
        
    
    
print("ev: ", ev)
print("deplasman: ", depl)
print("gol ev: ", goals_ev)
print("gol depl: ", goals_depl)
```

```python
session = requests.Session()

def get_rank(team, week):
    url = "https://www.tff.org/Default.aspx?pageID=1413&hafta=" + str(week-1) + "#grp"
    response = session.get(url)
    içerik = response.content
    soup = BeautifulSoup(içerik, "html.parser")
    str1 = "ctl00_MPane_m_1413_9321_ctnr_m_1413_9321_grvACetvel_ctl"
    str2 = "_lnkTakim"
    for i in range(1,22,1):
        if len(str(i)) == 1:
            team_id = str1 + "0" + str(i) + str2
        else:
            team_id = str1 + str(i) + str2
        result = soup.find_all("a",{"id": team_id})
        for x in result:
            if team in x.getText().split(".", 1)[1]:
                return x.getText().split(".")[0]
```

```python
session = requests.Session()

def get_last5_matches(team, week):
    first_week = week - 5
    url_str1 = "https://www.tff.org/Default.aspx?pageID=1501&hafta="
    url_str2 = "#grp"
    total_match = 0
    total_point = 0
    
    for x in range(first_week,week):
        url = url_str1 + str(x) + url_str2
        response = session.get(url)
        içerik = response.content
        soup = BeautifulSoup(içerik, "html.parser")
        str1 = "ctl00_MPane_m_1501_10169_ctnr_m_1501_10169_dtlHaftaninMaclari_ctl"
        str2 = "_Label"
        str3 = "_Labe4"
        for i in range(1,11,1):
            if len(str(i)) == 1:
                match_id = str1 + "0" + str(i) + str2
                goal_id = str1 + "0" + str(i) + str2
            else:
                match_id = str1 + str(i) + str2
                goal_id = str1 + str(i) + str2

            match_id1 = match_id + "4"
            match_id2 = match_id + "1"
            goal_id1 = goal_id + "5"
            goal_id2 = goal_id + "6"
            result_ev = soup.find_all("span",{"id": match_id1})
            result_depl = soup.find_all("span",{"id": match_id2})
            goal_1 = soup.find_all("span", {"id": goal_id1})
            goal_2 = soup.find_all("span", {"id": goal_id2})

            
            for t in goal_1:
                goal1 = int(t.getText())
            for t in goal_2:
                goal2 = int(t.getText())
            
            for t in result_ev:
                if team in t.getText():
                    total_match += 1
                    if goal1 > goal2:
                        total_point += 3
                    elif goal1 == goal2:
                        total_point += 1
                    
            for t in result_depl:
                if team in t.getText():
                    total_match += 1
                    if goal2 > goal1:
                        total_point += 3
                    elif goal1 == goal2:
                        total_point += 1
    
    print("Total Match: ", total_match, "  Total Point: ", total_point)
    return total_point/total_match
```

```python
def get_form(df, team, week):
    first_week = week - 5
    total_match = 0
    total_point = 0
    print("First Week: ", first_week, "Start Week: ", week)
    
    for w in range(first_week, week):
        df_temp = df[df["week"] == w]
        if (len(df_temp[df_temp["home"] == team]) > 0):
            total_match += 1
            goal1 = df_temp[df_temp["home"] == team]["goal1"][0]
            goal2 = df_temp[df_temp["home"] == team]["goal2"][0]
            if (goal1 > goal2):
                total_point += 3
            elif (goal1 == goal2):
                total_point += 1
        elif (len(df_temp[df_temp["away"] == team]) > 0):
            total_match += 1
            goal1 = df_temp[df_temp["away"] == team]["goal1"][0]
            goal2 = df_temp[df_temp["away"] == team]["goal2"][0]
            if (goal1 < goal2):
                total_point += 3
            elif (goal1 == goal2):
                total_point += 1
    
    print("Total Match: ", total_match, "  Total Point: ", total_point)
    return total_point/total_match
```
